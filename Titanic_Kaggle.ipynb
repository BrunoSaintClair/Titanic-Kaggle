```python
import pandas as pd
```

```python
df_treino = pd.read_csv('train.csv')
df_teste = pd.read_csv('test.csv')
```

```python
display(df_treino.info())
display(df_teste.info())
```

```python
df_treino = df_treino.loc[:,['PassengerId','Survived','Pclass','Sex','Age','SibSp','Parch']]
df_teste = df_teste.loc[:,['PassengerId','Pclass','Sex','Age','SibSp','Parch']]
```

```python
print('Treino:')
display(df_treino.isnull().sum())
print('--------')
print('Teste:')
display(df_teste.isnull().sum())
```

```python
media_idade = df_treino.groupby('Sex')['Age'].mean()

media_idade_mulheres = media_idade.iloc[0]
media_idade_homens = media_idade.iloc[1]

df_treino['Age'].fillna(
    df_treino['Sex'].apply(lambda x: media_idade_mulheres if x == 'female' else media_idade_homens),
    inplace=True)
```

```python
media_idade = df_teste.groupby('Sex')['Age'].mean()

media_idade_mulheres = media_idade.iloc[0]
media_idade_homens = media_idade.iloc[1]

df_teste['Age'].fillna(
    df_teste['Sex'].apply(lambda x: media_idade_mulheres if x == 'female' else media_idade_homens),
    inplace=True)
```

```python
display(df_treino.head())
display(df_teste.head())
```

```python
print('Treino:')
display(df_treino.isnull().sum())
print('--------')
print('Teste:')
display(df_teste.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()

df_treino['Sex'] = codificador.fit_transform(df_treino['Sex'])
df_teste['Sex'] = codificador.fit_transform(df_teste['Sex'])

display(df_treino.head())
display(df_teste.head())
```

```python
display(df_treino.info())
```

```python
display(df_teste.info())
```

```python
from sklearn.model_selection import train_test_split

X = df_treino.drop(['PassengerId','Survived'],axis=1)
y = df_treino.Survived

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
previsao_tree = clf.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf_knn = KNeighborsClassifier(n_neighbors=3)
clf_knn = clf_knn.fit(X_train, y_train)
previsao_knn = clf_knn.predict(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf_rl = LogisticRegression()
clf_rl = clf_rl.fit(X_train, y_train)
previsao_rl = clf_rl.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

score_tree = accuracy_score(y_test, previsao_tree)
score_knn = accuracy_score(y_test, previsao_knn)
score_rl = accuracy_score(y_test, previsao_rl)
```

```python
display(score_tree)
display(score_knn)
display(score_rl)
```

```python
X_test = df_teste.drop('PassengerId', axis=1)

previsao_rl_teste = clf_rl.predict(X_test)

df_teste['Survived'] = previsao_rl_teste

resultado = df_teste[['PassengerId', 'Survived']]
```

```python
resultado.to_csv('resultado1.csv', index=False)
```
